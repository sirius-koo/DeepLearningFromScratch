# **Chapter 01. 헬로 파이썬**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
plt.rc('font', family='NanumBarunGothic')
```

```python
# 1.5.2 넘파이 배열 생성하기
x = np.array([1.0, 2.0, 3.0])
print(x)
type(x)
```

```python
# 1.5.3 넘파이 산술 연산
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print(x + y)
print(x * y)
print(x / y)
print(x / 2.0)
```

```python
# 1.5.4 넘파이의 N차원 배열
A = np.array([[1, 2], [3, 4]])
B = np.array([[1, 2], [3, 4], [5, 6], [7, 8]]) # (4, 2)
C = np.array([[1, 2, 3], [4, 5, 6]]) # (2, 3)
print(A)
print(A.shape, B.shape, C.shape)
print(A.dtype)

B = np.array([[3, 0], [0, 6]])
print(A + B)
print(A * B)
print(A * 10)
```

```python
# 1.5.5 브로드캐스트
# 스칼라값이 연산하고자 하는 행렬의 형상과 같이 확대된 후 연산
A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])
A * B
```

```python
# 1.5.6 원소 접근
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
print(X[0])
print(X[0][1])
print()

# for문으로도 각 원소에 접근 가능
for row in X:
    print(row)
print()

# 평탄화
X = X.flatten()
print(X)
print(X[np.array([0, 2, 4])])
print()

print(X > 15)
print(X[X > 15])
```

```python
# 1.6 matplotlib
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)
# 그래프 그리기
plt.plot(x, y)
plt.show()
```

```python
# 1.6.2 pyplot의 기능
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") # cos 함수는 점선으로 그리기
plt.xlabel("나이") # x축 이름
plt.ylabel("성장점") # y축 이름
plt.title("sin & cos") # 제목
plt.legend() # 범례
plt.show()
```
